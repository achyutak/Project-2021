#Importing data


```python
docs = pd.DataFrame()
y = pd.DataFrame()
```

```python
for each_file in glob.glob("/content/drive/MyDrive/Project-2021/Sikka_AI/Saved_files/train_embeddings/train_word_embeddings_pickled_*.pkl"):
  df = pd.read_pickle(each_file)
  docs = docs.append(df)
for each_file in glob.glob("/content/drive/MyDrive/Project-2021/Sikka_AI/Saved_files/train_embeddings/train_labels_batch_*.pkl"):
  df = pd.read_pickle(each_file)
  y = y.append(df)
```

#Decision Tree Model

```python
execute = False
```

```python
if execute == True:
  print("Training the Decision Tree model")
  DT_model = DecisionTreeClassifier(max_depth=5,random_state=101)
  DT_model.fit(docs,y)
  print("Training Completed")

  # saving the model
  print("Saving the model..")
  with open('/content/drive/MyDrive/Project-2021/Sikka_AI/Saved_files/DT_model.pkl', 'wb') as f:
      pickle.dump(DT_model, f)
  print("Saving done!")
else:
  print("Loading saved Decision Tree Model..")
  with open('/content/drive/MyDrive/Project-2021/Sikka_AI/Saved_files/DT_model.pkl', 'rb') as f:
    DT_model = pickle.load(f)
  print("Loading done!")
```

```python
# x_test = pd.read_pickle('/content/drive/MyDrive/Project-2021/Sikka_AI/Saved_files/test.pkl')
# y_test = pd.read_pickle('/content/drive/MyDrive/Project-2021/Sikka_AI/Saved_files/test_labels.pkl')
```

```python
print("Training Accuracy : {}%".format(accuracy_score(y,DT_model.predict(docs))*100))
```

```python
for i in range(6):
  x_test_path = '/content/drive/MyDrive/Project-2021/Sikka_AI/Saved_files/test_embeddings/test_word_embeddings_pickled_' + str(i) + '.pkl'
  y_test_path = '/content/drive/MyDrive/Project-2021/Sikka_AI/Saved_files/test_embeddings/test_labels_batch_' + str(i) + '.pkl'
  x_test = pd.read_pickle(x_test_path)
  y_test = pd.read_pickle(y_test_path)
  print("Test Accuracy for batch {}: {}%".format(i,np.round(accuracy_score(y_test,DT_model.predict(x_test))*100,2)))

```
