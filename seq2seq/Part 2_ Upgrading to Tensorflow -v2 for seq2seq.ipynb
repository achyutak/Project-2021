````python
print("Renaming the files")
input_if_err = input("If any of the previous file upgrading lead to\
error, enter 1 else 0")
if int(input_if_err) == 1:
  file_name = input("Enter the name of the file with '-old' tag eg: mt_solver-old separated by ','")
  try:
    for i in file_name.split(','):
      src = i.strip() + ".py"
      dst = i.strip().split('-')[0] + '.py'
      os.rename(src,dst)
      print(src,'is renamed to',dst)
  except:
    print("Something is wrong try again")
````
